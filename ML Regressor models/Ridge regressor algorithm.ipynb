Y - WTMP column (Dependent variable) X - ATMP column (Independent variable)

```python
x=waves_data.ATMP.values.reshape(-1,1)
```

```python
y=waves_data.WTMP.values.reshape(-1,1)
```

Visualising the data using Matplotlib.pyplot

```python
plt.scatter(x,y,color='blue')
plt.xlabel('ATMP Semiconductor values')
plt.ylabel('WTMP Semiconductor values')
plt.title('Linear graph')
plt.show()
```

Splitting the dataset into training and test dataset using scikit learn

```python
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

Training the model

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import r2_score, mean_squared_error
model=Ridge()
```

In ridge, to find the alpha value. we can cross validate using grid search module

```python
parametres={'alpha':[0.001,0.01,0.1,1,5,10,12,15,20]}
```

```python
grid_model=GridSearchCV(model,parametres,scoring='neg_mean_squared_error',cv=5) #check those parametres in scikit documentation
grid_model.fit(x_train,y_train)
```

```python
grid_model.best_params_  #To find which parametrs was suitable 
```

```python
predictions=grid_model.predict(x_test)
predictions.flatten()
```

```python
score=r2_score(y_test,predictions)
print(score*100)
```

Drawing the best fit line in the graph

```python
plt.scatter(x,y,color='blue')
plt.plot(x_test,predictions,color='red')
plt.xlabel('ATMP Semiconductor values')
plt.ylabel('WTMP Semiconductor values')
plt.title('Linear graph')
plt.show()
```

As the values are scattered widely, the best fit line and predictions are done with ridge to avoid overfiiting
