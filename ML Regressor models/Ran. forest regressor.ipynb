```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
label_encode=LabelEncoder()
tips_data['sex_encode']=label_encode.fit_transform(tips_data['sex'])
tips_data['smoker_encode']=label_encode.fit_transform(tips_data['smoker'])
tips_data['day_encode']=label_encode.fit_transform(tips_data['day'])
tips_data['time_encode']=label_encode.fit_transform(tips_data['time'])
```

```python
x=tips_data[['total_bill','sex_encode','smoker_encode','day_encode','time_encode','size']]
x.head()
```

```python
y=tips_data.tip
y.head()
```

#Splitting the dataset into training and testing dataset

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

#Training the model

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
regressor=RandomForestRegressor()
```

```python
paramaters={'criterion':['squared_error','absolute_error','friedman_mse','poisson']}
```

```python
forest_regressor=GridSearchCV(regressor,paramaters,scoring='accuracy',cv=5)
forest_regressor.fit(x_train,y_train)
```

```python
forest_regressor.best_params_
```

```python
predictions=forest_regressor.predict(x_train)
predictions.flatten()
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_train,predictions)
print(score*100)
```
